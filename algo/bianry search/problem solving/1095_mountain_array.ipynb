```python
class Solution(object):
    def search(self, arr, target):
        peak = self.peakInMountainArray(arr)
        firstTry = self.orderAgnosticBS(arr, target, start=0, end=peak)
        
        if firstTry != -1:
            return firstTry
        # Try to search in the second half
        return self.orderAgnosticBS(arr, target, start=peak + 1, end=len(arr) - 1)

    def peakInMountainArray(self, arr):
        start = 0
        end = len(arr) - 1

        while start < end:
            mid = start + (end - start) // 2
            if arr[mid] > arr[mid + 1]:  # Peak found
                end = mid
            else:
                start = mid + 1
        
        # At the end, start == end and this is the peak of the mountain
        return start

    def orderAgnosticBS(self, arr, target, start, end):
        # Check if the array is ascending or descending
        isAsc = arr[start] < arr[end]
        
        while start <= end:
            mid = start + (end - start) // 2
            
            if arr[mid] == target:
                return mid
            
            if isAsc:
                if target > arr[mid]:
                    start = mid + 1
                else:
                    end = mid - 1
            else:  # Descending order case
                if target > arr[mid]:
                    end = mid - 1
                else:
                    start = mid + 1
                    
        return -1
```
