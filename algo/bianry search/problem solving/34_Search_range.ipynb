# Explanation of `searchRange` Method

The `searchRange` method finds the starting and ending positions of a given `target` value in a sorted array. If the target is not found, it returns `[-1, -1]`. This is done using a binary search approach, which is efficient with a time complexity of O(log n).

## Code Breakdown

### Method `searchRange`
- **Input**: 
  - `nums`: A list of integers sorted in non-decreasing order.
  - `target`: An integer that you want to find in the list.
- **Output**: 
  - A list of two integers representing the starting and ending positions of the `target` in the array. If the target is not found, it returns `[-1, -1]`.

### Step-by-Step Process

1. **Initialize Result**: 
   - We create a list `ans` initialized to `[-1, -1]`. This is the default result if the `target` is not found.

2. **First Occurrence Search**: 
   - Call the helper method `search` with `findStartIndex` set to `True` to find the first occurrence of the `target`.
   - Store the result in `ans[0]`.

3. **Second Occurrence Search**:
   - If the first occurrence was found (`ans[0]` is not `-1`), call the `search` method again with `findStartIndex` set to `False` to find the last occurrence of the `target`.
   - Store the result in `ans[1]`.

4. **Return Result**: 
   - Return `ans`, which now contains the indices of the first and last occurrences of the `target`.

### Method `search`
- This method performs a binary search to find the index of the `target`.
- **Parameters**:
  - `nums`: The list of integers.
  - `target`: The value to search for.
  - `findStartIndex`: A boolean indicating whether to find the first occurrence (`True`) or the last occurrence (`False`).

### Binary Search Logic

1. **Initialization**:
   - `start` is initialized to the beginning of the list (`0`).
   - `end` is initialized to the end of the list (`len(nums) - 1`).

2. **Binary Search Loop**:
   - While `start` is less than or equal to `end`:
     - Calculate `mid` as the average of `start` and `end`.
     - If `target` is greater than `nums[mid]`, it means the target is in the right half of the list, so update `start` to `mid + 1`.
     - If `target` is less than `nums[mid]`, it means the target is in the left half of the list, so update `end` to `mid - 1`.
     - If `target` is equal to `nums[mid]`, we have found a potential answer:
       - Store `mid` in `ans`.
       - If `findStartIndex` is `True`, move `end` to `mid - 1` to search for an earlier occurrence.
       - If `findStartIndex` is `False`, move `start` to `mid + 1` to search for a later occurrence.

3. **Return Index**:
   - Return `ans`, which contains the index of the first or last occurrence of `target` depending on `findStartIndex`.
