```python
from typing import List
# this will not work in duplicate values

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        pivot = self.findPivot(nums)
        
        # if we did not find a pivot, it means the array is not rotated
        if pivot == -1:
            return self.binarySearch(nums, target, start=0, end=len(nums) - 1)
        
        # if pivot is found, you have found 2 ascending sorted arrays
        # case 1: if the target is the pivot element
        if nums[pivot] == target:
            return pivot
        
        # case 2: if the target is in the left subarray
        if target >= nums[0]:
            return self.binarySearch(nums, target, start=0, end=pivot - 1)
        
        # case 3: if the target is in the right subarray
        return self.binarySearch(nums, target, start=pivot + 1, end=len(nums) - 1)
    
    def binarySearch(self, arr: List[int], target: int, start: int, end: int) -> int:
        while start <= end:
            mid = start + (end - start) // 2
            
            if arr[mid] == target:
                return mid
            elif target < arr[mid]:
                end = mid - 1
            else:
                start = mid + 1
        return -1
    
    def findPivot(self, arr: List[int]) -> int:
        start = 0
        end = len(arr) - 1
        
        while start <= end:
            mid = start + (end - start) // 2
            
            # 4 cases here
            
            # case 1: if mid is the pivot
            if mid < end and arr[mid] > arr[mid + 1]:
                return mid
            
            # case 2: if mid-1 is the pivot
            if mid > start and arr[mid] < arr[mid - 1]:
                return mid - 1
            
            # case 3: if left part is unsorted
            if arr[mid] <= arr[start]:
                end = mid - 1
            else:
                # case 4: if right part is unsorted
                start = mid + 1
                
        return -1
```
